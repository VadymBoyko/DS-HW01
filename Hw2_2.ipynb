```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Прочитайте файл 2017_jun_final.csv за допомогою методу read_csv
df = pd.read_csv("/content/sample_data/2017_jun_final.csv")
```

```python
#Прочитайте отриману таблицю, використовуючи метод head
print(df.head())
```

```python
#Визначте розмір таблиці за допомогою методу shape
print(df.shape)
```

```python
sns.barplot(data=df, x='Язык.программирования', y='Зарплата.в.месяц')
plt.xlabel('Мова програмування')
plt.ylabel('Зарплата')
plt.title('Залежність зарплати від мови програмування')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(data=df, y='Язык.программирования', order=df['Язык.программирования'].value_counts().index)
plt.ylabel('Мова програмування')
plt.xlabel('Кількість спеціалістів')
plt.title('Кількість спеціалістів по кожній мові програмування')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Город', y='Изменение.зарплаты.за.12.месяцев', data=df.sort_values('Город'))
plt.xlabel('Місто')
plt.ylabel('Зміна зарплати за 12 місяців')
plt.title('Динаміка зміни заробітної плати за містами')
plt.xticks(rotation=90)
plt.show()
```

```python
#Визначте типи всіх стовпців за допомогою dataframe.dtypes
print(df.dtypes)
```

```python
#Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
missing_ratio = df.isnull().sum() / len(df)
print(missing_ratio)
```

```python
#Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"
print(df)
df_nan=df.isnull().sum()
df_nan=df_nan[df_nan>0]
df=df.drop(columns=list(df_nan.index[1:]))
print(df)
```

```python
#Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"
missing_ratio_after_drop = df.isnull().sum() / len(df)
print(missing_ratio_after_drop)
```

```python
#Видаліть усі рядки у вихідній таблиці за допомогою методу dropna
df = df.dropna()
print(df)
```

```python
#Визначте новий розмір таблиці за допомогою методу shape
print(df.shape)
```

```python
#Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python
python_data = df[df["Язык.программирования"] == "Python"]
print(python_data)
```

```python
#Визначте розмір таблиці python_data за допомогою методу shape
print(python_data.shape)
```

```python
#Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
grouped_data = df.groupby("Должность")
```

```python
#Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg
# і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"
aggregated_data = grouped_data.agg({"Зарплата.в.месяц": ["min", "max"]})
print(aggregated_data)
```

```python
#Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць.
#Використовуйте її для методу apply та створіть новий стовпчик "avg"
def fill_avg_salary(row):
  return row.mean()
aggregated_data["avg"] = grouped_data["Зарплата.в.месяц"].apply(fill_avg_salary)
print(aggregated_data)
```

```python
#Створіть описову статистику за допомогою методу describe для нового стовпчика.
description = aggregated_data["avg"].describe()
print(description)
```

```python
#Збережіть отриману таблицю в CSV файл
aggregated_data.to_csv("Python_salary.csv", index=False)
```
