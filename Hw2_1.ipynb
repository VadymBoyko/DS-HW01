```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України"
url = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8'
df = pd.read_html(url, match='Коефіцієнт народжуваності в', decimal=',', thousands='.')[0]
print(df)
```

```python
#Вивести перші рядки таблиці за допомогою методу head
print(df.head())
```

```python
#Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
print(df.shape)
```

```python
#Замініть у таблиці значення "—" на значення NaN
df.replace('—', np.nan, inplace=True)
print(df)
```

```python
#Визначте типи всіх стовпців за допомогою dataframe.dtypes
print(df.dtypes)
```

```python
#Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"
df = df.astype(dict.fromkeys(['2014', '2019'], np.float64))
df.dtypes
```

```python
#Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
missing_ratio = df.isnull().sum() / len(df)
print(missing_ratio)
```

```python
#Видаліть з таблиці дані по всій країні, останній рядок таблиці
df = df.iloc[:-1]
print(df)
```

```python
#Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
df = df.fillna(df.mean())
print(df)
```

```python
#Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
regions_more_than_avg = df[df["2019"] > df["2019"].mean()]["Регіон"].tolist()
print(regions_more_than_avg)
```

```python
#У якому регіоні була найвища народжуваність у 2014 році?
print(df.loc[df["2014"].idxmax(), "Регіон"], df['2014'].max())
```

```python
#Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
df = df.set_index(df['Регіон'])
df["2019"].plot(kind='barh')
plt.title('Народжуваність по регіонах у 2019 році')
plt.ylabel("Регіон")
plt.grid()
```

```python
plt.plot(df.columns[1:] , df[df.columns[1:]].mean() , marker='o')
plt.xlabel('Рік')
plt.ylabel('Середній Коефіцієнт')
plt.title('Середній Коефіцієнт по рокам')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Регіон'], df['2000'], marker='o', label='2000')
plt.plot(df['Регіон'], df['2019'], marker='o', label='2019')
plt.xlabel('Регіон')
plt.ylabel('Коефіцієнт')
plt.title('Порівняння Коефіцієнт у 2000 та 2019 роках по регіонам')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
